```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pprint
import skimage.morphology as sk

sift = cv2.xfeatures2d.SIFT_create()
bf = cv2.BFMatcher()

template = cv2.imread('models/25.jpg',0)
query_bgr = cv2.imread('models/25.jpg')
train = cv2.imread('scenes/m5.png',0)
train_rgb = cv2.imread('scenes/m5.png')
#train = template, scale function will be close to 1 if you use real baricenter

#fare per ogni query
kp_template, des_template = sift.detectAndCompute(template, None)
kp_train, des_train = sift.detectAndCompute(train, None)

matches = bf.knnMatch(des_train, des_template, k=2)
good_matches = []
for m, n in matches:
    if m.distance < 0.7 * n.distance:
        good_matches.append(m)

good_matches = tuple(good_matches)
sum_of_coordinates = [0,0]

for i in kp_template:
    sum_of_coordinates = [sum_of_coordinates[0]+i.pt[0],sum_of_coordinates[1]+i.pt[1]]

baricenter = [sum_of_coordinates[0]/len(kp_template),sum_of_coordinates[1]/len(kp_template)]
```

## Geometry Validation:

```python
#fig = plt.figure()
#ax = plt.axes(projection='3d')
#ax.view_init(45,45)

scale_ratios = []
rotations = []
x_center = []
y_center = []

#fare per ogni query
for m in good_matches:

    query_center = baricenter
    train_kp = kp_train[m.queryIdx]
    query_kp = kp_template[m.trainIdx]
    
    query_vector = np.array(query_kp.pt)-np.array(query_center)
    
    scale_kp = train_kp.size / query_kp.size
    scale_ratios.append(scale_kp)
    
    rot_kp = train_kp.angle - query_kp.angle
    rotations.append(rot_kp)
    
    train_vector_x = train_kp.pt[0] + scale_kp * (np.cos(rot_kp) * query_vector[0] - np.sin(rot_kp) * query_vector[1])
    x_center.append(train_vector_x)
    
    train_vector_y = train_kp.pt[1] + scale_kp * (np.sin(rot_kp) * query_vector[0] + np.cos(rot_kp) * query_vector[1])
    y_center.append(train_vector_y) 
    
    

counts_scale, bins_scale, _ = plt.hist(scale_ratios, bins='auto')
#plt.show()

#plt.hist(rotations, bins='auto')
#plt.show()

# scale più probabile dell'immagine, serve per la grandezza delle regioni 
# valutare se usare la media
scale = bins_scale[np.argmax(counts_scale)]
    
x_bin_size = scale * template.shape[1] / 4
y_bin_size = scale * template.shape[0] / 4
x_bins = int(np.ceil(train.shape[1] / x_bin_size)+2)
y_bins = int(np.ceil(train.shape[0] / y_bin_size)+2)


accumulator_matches = {}
accumulator_mask = np.zeros((x_bins, y_bins))
for i,m in enumerate(good_matches):
    x = int(np.floor(x_center[i]/(x_bin_size)))
    y = int(np.floor(y_center[i]/(y_bin_size)))
    s = scale_ratios[i]
    r = rotations[i]
    accumulator_matches[(x,y)] = accumulator_matches.get((x,y), [])
    accumulator_matches[(x,y)].append(m)
    if x>0 and y>0 and x< x_bins and y<y_bins:
        accumulator_mask[x,y] += 1

mask = sk.local_maxima(accumulator_mask)
accumulator_mask[mask != 1]=0
accumulator_mask[accumulator_mask<10]=0
#pprint.pprint(accumulator_matches)

# andrà fatto tante volte quante scatole troviamo
    
for i in accumulator_matches.keys():
    if i[0]>=0 and i[1]>=0 and i[1]<17 and i[0]<17:
        if (accumulator_mask[i[0],i[1]]>0):
            
            print(accumulator_mask[i[0],i[1]])
            print(i[0],i[1])
            src_pts = []
            dst_pts = []

            for m in accumulator_matches[i]:
                src_pts.append(kp_template[m.trainIdx].pt)
                dst_pts.append(kp_train[m.queryIdx].pt)

            M, _ = cv2.findHomography(np.float32(src_pts).reshape(-1,1,2),np.float32(dst_pts).reshape(-1,1,2),cv2.RANSAC,5.0)

            h, w, d = query_bgr.shape
            pts = np.float32([[0, 0], [0, h - 1], [w - 1, h - 1], [w - 1, 0]]).reshape(-1,1,2)
            print(src_pts)
            print(pts)
            dst = cv2.perspectiveTransform(pts, M).reshape(-1,2)
            print(dst)
            
            final_image = cv2.polylines(train_rgb, [np.int32(dst)], True, (0, 255, 0), 3, cv2.LINE_AA)

            plt.imshow(final_image)
            plt.show()
```
